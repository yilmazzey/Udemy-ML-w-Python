```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel("Iris.xls")
data.head()
```

```python
x=data.drop(columns=["iris"])
y=data["iris"]
iris_types=data["iris"].unique()
print(iris_types)
```

```python
#distribution of iris types along different variables
setosa=data[data["iris"]=="Iris-setosa"]
versicolor=data[data["iris"]=="Iris-versicolor"]
virginica=data[data["iris"]=="Iris-virginica"]

# Sepal size
plt.figure(figsize=(12,8))
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(setosa['sepal length'], setosa['sepal width'], label='Setosa', c='red')
plt.scatter(versicolor['sepal length'], versicolor['sepal width'], label='Versicolor', c='blue')
plt.scatter(virginica['sepal length'], virginica['sepal width'], label='Virginica', c='green')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend()

# Petal size
plt.subplot(1, 2, 2)
plt.scatter(setosa['petal length'], setosa['petal width'], label='Setosa', c='red')
plt.scatter(versicolor['petal length'], versicolor['petal width'], label='Versicolor', c='blue')
plt.scatter(virginica['petal length'], virginica['petal width'], label='Virginica', c='green')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.legend()

plt.tight_layout()
plt.show()

```

train test split with cross validation

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=42)
```

Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

classsification algorithms(logistic regression, knn, svm, naive bayes, desicion tree, random forest)

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
logr=LogisticRegression(random_state=42)
logr.fit(X_train,y_train)

y_pred=logr.predict(X_test)

#confusion matrix to evaluate the results
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
#k-nearest neighbours
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier(n_neighbors=2,metric="minkowski")
knn.fit(X_train,y_train)

y_pred=knn.predict(X_test)
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
#svm
from sklearn.svm import SVC
svc=SVC(kernel="poly")
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
#naive bayes
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
gnb.fit(X_train,y_train)
y_pred =gnb.predict(X_test)
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
#desicion tree
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier(criterion="entropy")
dtc.fit(X_train,y_train)
y_pred=dtc.predict(X_test)
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
#random forest
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=10,criterion="entropy")
rfc.fit(X_train,y_train)
y_pred=rfc.predict(X_test)
cm=confusion_matrix(y_test,y_pred)
print(cm)
```
